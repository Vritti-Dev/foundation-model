## What you'll build

A neural network can't read text. It only does arithmetic on numbers.
So the very first thing any language model needs is a way to turn a
string like `"hello"` into a list of integers, and back again.
That translator is called a **tokenizer**.

In this warm-up you'll write the simplest possible tokenizer: one that
treats every single **character** as one token. The letter `h` becomes
some number, `e` becomes another, and so on. No clever splitting into
words or sub-words yet -- just characters.

Here is the plan, which is exactly what your `CharTokenizer` will do:

1. Look at all the text you're given and collect every **unique**
   character. Sort them so the order is always the same. That sorted
   list is your **vocabulary**.
2. Build two lookup tables: `stoi` ("string to integer") maps each
   character to its position in the vocabulary, and `itos` ("integer
   to string") maps back.
3. `encode` walks through a string and replaces each character with its
   id. `decode` does the reverse.

The reason we sort the vocabulary is **determinism**: given the same
text, you always get the same ids. That matters later when we want a
model trained today to behave the same way tomorrow.

```
# Reference solution (single source of truth: reference/tokenizer/char_tokenizer.py)

"""A minimal character-level tokenizer.

Maps each unique character in a body of text to an integer id and back.
This is the simplest possible tokenizer: one token per character, no
merging or special handling. Great for learning how text turns into the
integer sequences a language model actually consumes.
"""


class CharTokenizer:
    """Build a fixed vocabulary from `text` and convert chars <-> ids.

    The vocabulary is the sorted set of characters seen in `text`, so ids
    are assigned deterministically (id 0 is the lexicographically smallest
    character, and so on).
    """

    def __init__(self, text):
        # Sorted, de-duplicated characters form the stable vocabulary.
        self.vocab = sorted(set(text))
        # string-to-index and index-to-string lookup tables.
        self.stoi = {ch: i for i, ch in enumerate(self.vocab)}
        self.itos = {i: ch for i, ch in enumerate(self.vocab)}
        self.vocab_size = len(self.vocab)

    def encode(self, s):
        """Turn a string into a list of integer token ids."""
        return [self.stoi[ch] for ch in s]

    def decode(self, ids):
        """Turn a list of integer token ids back into a string."""
        return "".join(self.itos[i] for i in ids)
```
