# Module 10 - What's Next: How Real LLMs Differ

You've built a working GPT from first principles. The frontier models you
hear about -- the ones that write essays and code -- are the *same idea*
you just implemented, scaled and refined along a few key axes. Here's
what changes between your tiny Shakespeare model and a production LLM.

## 1. Scale

Your model has under a million parameters. Frontier models have hundreds
of billions to trillions. The architecture is nearly identical -- more
layers, wider embeddings, more heads, a longer context window. Scaling
laws showed that loss falls predictably as you grow model size, data,
and compute together, which is why the industry kept making models
bigger. Training one run can cost millions of dollars and weeks on
thousands of GPUs.

## 2. Data

You trained on a few KB of Shakespeare. Real models train on trillions
of tokens scraped from the web, books, and code, then heavily filtered
and deduplicated. They also use **subword tokenizers** (BPE) instead of
your character-level one, so each token carries more meaning and
sequences are shorter. Data quality and mixture matter as much as
quantity.

## 3. Fine-tuning

A raw pretrained model just predicts the next token -- it doesn't
naturally follow instructions. **Supervised fine-tuning (SFT)** continues
training on curated (instruction, response) pairs so the model learns to
be a helpful assistant rather than an autocomplete engine.

## 4. RLHF and alignment

**Reinforcement Learning from Human Feedback** goes further: humans rank
model responses, a reward model learns those preferences, and the LLM is
optimised to produce answers people prefer -- more helpful, honest, and
harmless. Newer variants (DPO, RLAIF, constitutional methods) chase the
same goal of aligning model behaviour with human intent.

## 5. Inference tricks

Production systems add KV caching for speed, quantization to shrink
models, better sampling (temperature, top-p) instead of greedy decoding,
and long-context techniques. The core forward pass, though, is the one
you wrote.

## Where to go from here

* Read Karpathy's *nanoGPT* and *makemore* -- the spiritual parents of
  this course.
* Try a BPE tokenizer and compare loss per character.
* Scale your capstone model up on a bigger corpus and longer context.
* Explore SFT: fine-tune on instruction data and see the behaviour shift.

You now understand, end to end, what an LLM actually is. That intuition
doesn't go out of date.
