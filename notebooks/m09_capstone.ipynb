### Picking a corpus

* Aim for at least a few hundred KB of text -- more is better.
* It must be a **character-level** friendly plain-text file (`.txt`).
* Keep it legal and appropriate: use public-domain text (Project
  Gutenberg is a goldmine) or text you own.

On Colab, upload your file with the snippet below, or paste text
directly into the `learner_corpus` string.

```
# Bootstrap (run me first / safe to re-run): clone the course repo on
# Colab and put it on sys.path so `from reference...` and `from grader...`
# work in every cell below. No-op when running locally.
import os, sys
try:
    import google.colab  # noqa: F401
    _repo_dir = '/content/foundation-model'
    if not os.path.exists(_repo_dir):
        !git clone -q https://github.com/Vritti-Dev/foundation-model.git {_repo_dir}
    if _repo_dir not in sys.path:
        sys.path.insert(0, _repo_dir)
    os.chdir(_repo_dir)
    print('course repo ready:', _repo_dir)
except ImportError:
    pass  # local: cwd already on sys.path

import torch
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('device:', device)
```

```
# --- Provide YOUR corpus here ---
# Option A: paste/keep text directly in this string.
learner_corpus = (
    'To be, or not to be, that is the question:\n'
    'Whether tis nobler in the mind to suffer\n'
    'The slings and arrows of outrageous fortune,\n'
) * 40

# Option B (Colab): upload a .txt file and load it instead.
#   from google.colab import files
#   up = files.upload()
#   learner_corpus = open(next(iter(up)), encoding='utf-8').read()

assert len(learner_corpus) > 200, 'corpus is very small; add more text'
print('corpus chars:', len(learner_corpus))
```

```
# Train on your corpus.
from reference.gpt.train import train, make_submission_token
import torch

max_iters = 1000 if torch.cuda.is_available() else 100
config = 't4' if torch.cuda.is_available() else 'cpu'
res = train(config=config, max_iters=max_iters, seed=1337, data=learner_corpus)
print(f"initial loss {res['initial_loss']:.4f} -> final loss {res['final_loss']:.4f}")
```

```
# (Optional) Save a checkpoint you can download from Colab.
import torch
torch.save(res['model'].state_dict(), 'capstone_model.pt')
print('saved capstone_model.pt')
# On Colab, download it with:
#   from google.colab import files; files.download('capstone_model.pt')
```
