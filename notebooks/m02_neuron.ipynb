## What you'll build

A neural network is built out of one tiny repeated unit: the **neuron**.
A neuron takes some inputs, multiplies each by a **weight**, adds them
up, adds a **bias**, and squashes the result through a nonlinearity.
That's it. Stack thousands of these and you get a language model.

In this module you'll do two things by hand:

1. **Forward pass.** Compute a single neuron's output:
   `y = tanh(w . x + b)`. The `tanh` keeps the output between -1 and 1.
2. **One gradient step.** We'll define a simple loss -- how far the
   neuron's output is from a target -- and nudge the weights in the
   direction that *reduces* that loss. This is the heartbeat of all
   training: measure error, compute the slope (gradient) of the error
   with respect to each weight, take a small step downhill.

fact: for loss `L = (y - target)^2` with `y = tanh(z)` and `z = w.x + b`,
the chain rule gives `dL/dw = 2*(y - target) * (1 - y^2) * x`. After one
step of `w <- w - lr * dL/dw`, the loss should go **down**.

## Reveal the reference solution

Run the hidden cell to see the forward pass and the hand-derived
gradient step.

```
# Reference solution (self-contained: re-declares the inputs too).
import numpy as np
x = np.array([0.5, -1.0, 2.0])
w = np.array([0.1, 0.2, -0.1])
b = 0.0
target = 1.0
lr = 0.1
z = w @ x + b
y = np.tanh(z)
loss_before = (y - target) ** 2
grad_w = 2 * (y - target) * (1 - y ** 2) * x
w_new = w - lr * grad_w
y_after = np.tanh(w_new @ x + b)
loss_after = (y_after - target) ** 2
```
